## FRQ Question 2


Consider a guessing game in which a player tries to guess a hidden word. The hidden word contains only capital
letters and has a length known to the player. A guess contains only capital letters and has the same length as the
hidden word.

After a guess is made, the player is given a hint that is based on a comparison between the hidden word and the
guess. Each position in the hint contains a character that corresponds to the letter in the same position in the
guess. The following rules determine the characters that appear in the hint.

    </tr>

        <td>not in the hidden word, </td>
        <td>"*" </td>
    </tr>
<table>

The HiddenWord class will be used to represent the hidden word in the game. The hidden word is passed to
the constructor. The class contains a method, getHint, that takes a guess and produces a hint. 

For example, suppose the variable puzzle is declared as follows. 

HiddenWord puzzle = new HiddenWord("HARPS");

Write the complete HiddenWord class, including any necessary instance variables, its constructor, and the
method, getHint, described above. You may assume that the length of the guess is the same as the length of
the hidden word. 


````java
// CODE_RUNNER: HiddenWord
public class Main {

    // ===============================
    // 2015 FRQ #2 — HiddenWord class
    // ===============================
    public static class HiddenWord {

        // STUDENT CODE STARTS HERE
        private String word;
        // STUDENT CODE ENDS HERE

        /**
         * Constructs a HiddenWord object.
         * @param hWord the word to be guessed
         * Precondition: hWord contains only uppercase letters.
         */
        public HiddenWord(String hWord) {

            // STUDENT CODE STARTS HERE
            word = hWord;
            // STUDENT CODE ENDS HERE
        }

        /**
         * Returns a hint for guess.
         * @param guess the word guessed
         * Precondition: guess has the same length as the hidden word
         *               and contains only uppercase letters.
         */
        public String getHint(String guess) {

            // STUDENT CODE STARTS HERE

            String hint = "";

            for (int i = 0; i < guess.length(); i++) {
                String guessLetter = guess.substring(i, i + 1);

                if (guessLetter.equals(word.substring(i, i + 1))) {
                    hint += guessLetter;
                } else if (word.indexOf(guessLetter) != -1) {
                    hint += "+";
                } else {
                    hint += "*";
                }
            }

            return hint;

            // STUDENT CODE ENDS HERE
        }
    }

    // ===============================
    // Driver to prove it runs
    // ===============================
    public static void main(String[] args) {

        HiddenWord puzzle = new HiddenWord("HARPS");

        System.out.println(puzzle.getHint("AAAAA")); // +****
        System.out.println(puzzle.getHint("HELLO")); // H****
        System.out.println(puzzle.getHint("HEART")); // H*++*
        System.out.println(puzzle.getHint("HARMS")); // HAR*S
    }
}

// Required by runner
Main.main(null);
````

## Interactive HiddenWord Game

Play this mini-game to experience the `getHint()` method in action! The game implements the same logic you need to code for the FRQ.

**How it works:**
- You have 6 attempts to guess the 5-letter word
- Green letters are in the correct position (same as returning the letter)
- Yellow letters are in the word but wrong position (same as returning `+`)
- Gray letters are not in the word (same as returning `*`)

````java
%%html
<style>
  .hw-game-container { font-family: Arial, sans-serif; max-width: 350px; margin: 20px auto; text-align: center; }
  .hw-grid { display: grid; grid-template-rows: repeat(6, 1fr); gap: 5px; margin-bottom: 20px; }
  .hw-row { display: grid; grid-template-columns: repeat(5, 1fr); gap: 5px; }
  .hw-cell { width: 50px; height: 50px; border: 2px solid #d3d6da; display: flex; align-items: center; justify-content: center; font-size: 24px; font-weight: bold; text-transform: uppercase; }
  .hw-cell.correct { background-color: #6aaa64; color: white; border-color: #6aaa64; }
  .hw-cell.present { background-color: #c9b458; color: white; border-color: #c9b458; }
  .hw-cell.absent { background-color: #787c7e; color: white; border-color: #787c7e; }
  .hw-input-area { margin: 15px 0; }
  .hw-input { font-size: 18px; padding: 10px; width: 150px; text-transform: uppercase; text-align: center; }
  .hw-btn { font-size: 16px; padding: 10px 20px; margin: 5px; cursor: pointer; background-color: #4CAF50; color: white; border: none; border-radius: 5px; }
  .hw-btn:hover { background-color: #45a049; }
  .hw-btn.reset { background-color: #2196F3; }
  .hw-message { font-size: 16px; margin-top: 10px; min-height: 24px; }
  .hw-legend { margin-top: 15px; font-size: 14px; text-align: left; background: #f5f5f5; padding: 10px; border-radius: 5px; color: black; }
  .hw-legend-item { display: flex; align-items: center; margin: 5px 0; }
  .hw-legend-box { width: 20px; height: 20px; margin-right: 10px; }
</style>

<div class="hw-game-container">
  <h3>HiddenWord Game</h3>
  <div id="hwGrid" class="hw-grid"></div>
  <div class="hw-input-area">
    <input type="text" id="hwInput" class="hw-input" maxlength="5" placeholder="GUESS">
    <br>
    <button class="hw-btn" onclick="submitHwGuess()">Submit</button>
    <button class="hw-btn reset" onclick="resetHwGame()">New Game</button>
  </div>
  <div id="hwMessage" class="hw-message"></div>
  <div class="hw-legend">
    <strong>Legend (Same as FRQ getHint!):</strong>
    <div class="hw-legend-item"><div class="hw-legend-box" style="background:#6aaa64;"></div>Letter in correct position (returns letter)</div>
    <div class="hw-legend-item"><div class="hw-legend-box" style="background:#c9b458;"></div>Letter in word, wrong position (returns +)</div>
    <div class="hw-legend-item"><div class="hw-legend-box" style="background:#787c7e;"></div>Letter not in word (returns *)</div>
  </div>
</div>

<script>
// HiddenWord Game - Implements the FRQ getHint() logic
const hwWords = ['HARPS', 'GRAPE', 'CLASH', 'BRAIN', 'PLANT', 'STORM', 'LIGHT', 'SPACE', 'DREAM', 'WORLD'];
let hwTargetWord = '';
let hwCurrentRow = 0;
let hwGameOver = false;

function initHwGame() {
  hwTargetWord = hwWords[Math.floor(Math.random() * hwWords.length)];
  hwCurrentRow = 0;
  hwGameOver = false;
  document.getElementById('hwMessage').textContent = '';
  document.getElementById('hwInput').value = '';
  
  const grid = document.getElementById('hwGrid');
  grid.innerHTML = '';
  for (let r = 0; r < 6; r++) {
    const row = document.createElement('div');
    row.className = 'hw-row';
    for (let c = 0; c < 5; c++) {
      const cell = document.createElement('div');
      cell.className = 'hw-cell';
      cell.id = `hw-cell-${r}-${c}`;
      row.appendChild(cell);
    }
    grid.appendChild(row);
  }
}

// This is the SAME LOGIC as the FRQ getHint() method!
function hwGetHint(guess) {
  let result = [];
  for (let i = 0; i < guess.length; i++) {
    const guessLetter = guess.charAt(i);
    if (guessLetter === hwTargetWord.charAt(i)) {
      result.push({ letter: guessLetter, status: 'correct' }); // Same position = return letter
    } else if (hwTargetWord.indexOf(guessLetter) !== -1) {
      result.push({ letter: guessLetter, status: 'present' }); // In word = return +
    } else {
      result.push({ letter: guessLetter, status: 'absent' }); // Not in word = return *
    }
  }
  return result;
}

function submitHwGuess() {
  if (hwGameOver) return;
  const input = document.getElementById('hwInput');
  const guess = input.value.toUpperCase();
  
  if (guess.length !== 5) {
    document.getElementById('hwMessage').textContent = 'Please enter a 5-letter word!';
    return;
  }
  if (!/^[A-Z]+$/.test(guess)) {
    document.getElementById('hwMessage').textContent = 'Only letters allowed!';
    return;
  }
  
  const hint = hwGetHint(guess);
  for (let i = 0; i < 5; i++) {
    const cell = document.getElementById(`hw-cell-${hwCurrentRow}-${i}`);
    cell.textContent = hint[i].letter;
    cell.classList.add(hint[i].status);
  }
  
  if (guess === hwTargetWord) {
    document.getElementById('hwMessage').textContent = 'You won! The getHint() returned the full word!';
    hwGameOver = true;
  } else {
    hwCurrentRow++;
    if (hwCurrentRow >= 6) {
      document.getElementById('hwMessage').textContent = `Game Over! The word was: ${hwTargetWord}`;
      hwGameOver = true;
    }
  }
  input.value = '';
}

function resetHwGame() { initHwGame(); }

// Allow Enter key to submit
document.addEventListener('DOMContentLoaded', function() {
  const input = document.getElementById('hwInput');
  if (input) {
    input.addEventListener('keypress', function(e) {
      if (e.key === 'Enter') submitHwGuess();
    });
  }
});

initHwGame();
</script>
````
